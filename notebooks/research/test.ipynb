#### Merging all the dataframes


```python
import pandas as pd
import os

# List all the files to be merged
file_list = [
    'final_scrape/dados_detalhados_olx_1.csv',
    'final_scrape/dados_detalhados_olx_2.csv',
    'final_scrape/dados_detalhados_olx_3.csv',
    'final_scrape/dados_detalhados_olx_4.csv',
]

# Create an empty list to store individual dataframes
df_list = []

# Iterate through each file, read it, and append to the list
for file_name in file_list:
    if os.path.exists(file_name):
        df = pd.read_csv(file_name)
        df_list.append(df)
    else:
        print(f"File {file_name} not found!")

# Concatenate all dataframes in the list
final_dataframe = pd.concat(df_list, axis=0, ignore_index=True)


# Save the final dataframe to CSV
final_dataframe.to_csv('final_dataframe.csv', index=False)

print("Dataframes merged and saved to 'final_dataframe.csv'")
```
