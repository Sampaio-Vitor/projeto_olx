## SCRAPE DE DADOS GERAIS

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import time

def fetch_data(driver, url):
    driver.get(url)
    WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'h2[class="horizontal title sc-ifAKCX hUnWqk"]')))
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    titles = soup.find_all('h2', attrs={'class': 'horizontal title sc-ifAKCX hUnWqk'})
    links = soup.find_all('a', attrs={'data-ds-component': 'DS-NewAdCard-Link'})

    data = []

    for title, link in zip(titles, links):
        row = {
            'titulo': title.text,
            'link': link['href'],
            'date': datetime.now()
        }

        data.append(row)

    return data

# Configurando as opções do WebDriver
options = Options()
options.add_argument("--start-maximized")

# Coleta dados de todas as páginas disponíveis
page = 1

driver = webdriver.Chrome(options=options)

while True:
    if page == 1:
        url = 'https://www.olx.com.br/imoveis/venda/apartamentos/estado-mg/belo-horizonte-e-regiao/'
    else:
        url = f'https://www.olx.com.br/imoveis/venda/apartamentos/estado-mg/belo-horizonte-e-regiao?o={page}'

    new_data = fetch_data(driver, url)
    if not new_data:  # Pára se a página não contiver mais dados
        break

    # Adicionando os dados coletados em um DataFrame
    df = pd.DataFrame(new_data)
    print(f"Page number: {page}")
    print(f"Length of DataFrame: {len(df)}")
    print(f"Sample data:\n{df.sample(10)}")

    # Salvando dados no CSV
    if page == 1:
        df.to_csv('dados_olx.csv', index=False)
    else:
        df.to_csv('dados_olx.csv', mode='a', header=False, index=False)
        
    page += 1
    time.sleep(15)  # Pausa por 15 segundos entre as solicitações

driver.quit()
```

## SCRAPE DE DADOS POR ZONA


```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import time

def fetch_data(driver, url):
    driver.get(url)
    WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'h2[class="horizontal title sc-ifAKCX hUnWqk"]')))
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    titles = soup.find_all('h2', attrs={'class': 'horizontal title sc-ifAKCX hUnWqk'})
    links = soup.find_all('a', attrs={'data-ds-component': 'DS-NewAdCard-Link'})

    data = []

    for title, link in zip(titles, links):
        row = {
            'titulo': title.text,
            'link': link['href'],
            'date': datetime.now()
        }

        data.append(row)

    return data

# Configurando as opções do WebDriver
options = Options()
options.add_argument("--start-maximized")

# Lista de regiões
regions = ['zona-oeste']

driver = webdriver.Chrome(options=options)

for region in regions:
    # Coleta dados de todas as páginas disponíveis
    page = 1
    while True:
        if page == 1:
            url = f'https://www.olx.com.br/imoveis/venda/apartamentos/estado-mg/belo-horizonte-e-regiao/{region}'
        else:
            url = f'https://www.olx.com.br/imoveis/venda/apartamentos/estado-mg/belo-horizonte-e-regiao/{region}?o={page}'

        new_data = fetch_data(driver, url)
        if not new_data:  # Pára se a página não contiver mais dados
            break

        # Adicionando os dados coletados em um DataFrame
        df = pd.DataFrame(new_data)
        print(f"Region: {region}")
        print(f"Page number: {page}")
        print(f"Length of DataFrame: {len(df)}")
        print(f"Sample data:\n{df.sample(10)}")

        # Salvando dados no CSV
        if page == 1:
            df.to_csv(f'links_{region}.csv', index=False)
        else:
            df.to_csv(f'links_{region}.csv', mode='a', header=False, index=False)
            
        page += 1
        time.sleep(3)  # Pausa por 3 segundos entre as solicitações

driver.quit()
```
