```python
from bayes_opt import BayesianOptimization
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
import numpy as np

# Define the objective function to be maximized (negative MSE)
def objective(learning_rate, n_estimators, max_depth):
    learning_rate = 10 ** learning_rate
    n_estimators = int(n_estimators)
    max_depth = int(max_depth)

    # Instantiate the XGBRegressor with the current hyperparameters
    regressor = XGBRegressor(learning_rate=learning_rate, n_estimators=n_estimators, max_depth=max_depth, random_state=42)

    # Train the regressor
    regressor.fit(X_train, y_train)

    # Make predictions on the validation data
    y_pred = regressor.predict(X_test)

    # Calculate the negative MSE (BayesianOptimization minimizes the objective)
    mse = -mean_squared_error(y_test, y_pred)
    return mse

# Define the parameter bounds for Bayesian optimization
pbounds = {
    'learning_rate': (-5, 0),  # log scale: 10^-5 to 1
    'n_estimators': (100, 500),  # discrete values between 100 and 500
    'max_depth': (3, 10),  # discrete values between 3 and 10
}

# Instantiate the BayesianOptimization object
optimizer = BayesianOptimization(f=objective, pbounds=pbounds, random_state=42)

# Perform the Bayesian optimization
optimizer.maximize(init_points=10, n_iter=50)

# Get the best hyperparameters
best_params = optimizer.max['params']
best_learning_rate = 10 ** best_params['learning_rate']
best_n_estimators = int(best_params['n_estimators'])
best_max_depth = int(best_params['max_depth'])

# Instantiate the best regressor
best_regressor = XGBRegressor(learning_rate=best_learning_rate, n_estimators=best_n_estimators, max_depth=best_max_depth, random_state=42)


```
